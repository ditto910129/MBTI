```python
import pandas as pd
import numpy as np
```

```python
ENFJ = pd.read_csv('./데이터 수집/MBTI별 디시인사이드 단어사용/결과/ENFP 디시인사이드 단어사용.csv', encoding='cp949')
SF = pd.read_csv('./데이터 수집/장르별 영화 평점/결과/SF 네이버 영화 단어사용.csv', encoding='cp949')
공포 = pd.read_csv('./데이터 수집/장르별 영화 평점/결과/공포 네이버 영화 단어사용.csv', encoding='cp949')
드라마 = pd.read_csv('./데이터 수집/장르별 영화 평점/결과/드라마 네이버 영화 단어사용.csv', encoding='cp949')
로맨스 = pd.read_csv('./데이터 수집/장르별 영화 평점/결과/로맨스 네이버 영화 단어사용.csv', encoding='cp949')
스릴러 = pd.read_csv('./데이터 수집/장르별 영화 평점/결과/스릴러 네이버 영화 단어사용.csv', encoding='cp949')
액션 = pd.read_csv('./데이터 수집/장르별 영화 평점/결과/액션 네이버 영화 단어사용.csv', encoding='cp949')
코메디 = pd.read_csv('./데이터 수집/장르별 영화 평점/결과/코메디 네이버 영화 단어사용.csv', encoding='cp949')
판타지 = pd.read_csv('./데이터 수집/장르별 영화 평점/결과/판타지 네이버 영화 단어사용.csv', encoding='cp949')
```

```python
ENFJ_list = ENFJ['word'].values.tolist()
SF_list = SF['word'].values.tolist()
공포_list = 공포['word'].values.tolist()
드라마_list = 드라마['word'].values.tolist()
로맨스_list = 로맨스['word'].values.tolist()
스릴러_list = 스릴러['word'].values.tolist()
액션_list = 액션['word'].values.tolist()
코메디_list = 코메디['word'].values.tolist()
판타지_list = 판타지['word'].values.tolist()
```

```python
ENFJ_sum = " ".join(ENFJ_list)
SF_sum = " ".join(SF_list)
공포_sum = " ".join(공포_list)
드라마_sum = " ".join(드라마_list)
로맨스_sum = " ".join(로맨스_list)
스릴러_sum = " ".join(스릴러_list)
액션_sum = " ".join(액션_list)
코메디_sum = " ".join(코메디_list)
판타지_sum = " ".join(판타지_list)
```

```python
doc_list = []
```

```python
doc_list.append(ENFJ_sum)
doc_list.append(SF_sum)
doc_list.append(공포_sum)
doc_list.append(드라마_sum)
doc_list.append(로맨스_sum)
doc_list.append(스릴러_sum)
doc_list.append(액션_sum)
doc_list.append(코메디_sum)
doc_list.append(판타지_sum)
```

```python
doc_list
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect_simple = TfidfVectorizer()
feature_vect_simple = tfidf_vect_simple.fit_transform(doc_list)

print(feature_vect_simple.shape)
print(type(feature_vect_simple))
```

```python
feature_vect_dense = feature_vect_simple.todense()

ENFJ_최종 = np.array(feature_vect_dense[0].reshape(-1,))
SF_최종 = np.array(feature_vect_dense[1].reshape(-1,))
공포_최종 = np.array(feature_vect_dense[2].reshape(-1,))
드라마_최종 = np.array(feature_vect_dense[3].reshape(-1,))
로맨스_최종 = np.array(feature_vect_dense[4].reshape(-1,))
스릴러_최종 = np.array(feature_vect_dense[5].reshape(-1,))
액션_최종 = np.array(feature_vect_dense[6].reshape(-1,))
코메디_최종 = np.array(feature_vect_dense[7].reshape(-1,))
판타지_최종 = np.array(feature_vect_dense[8].reshape(-1,))
```

```python
from sklearn.metrics.pairwise import cosine_similarity

print('ENFJ와 SF와의 유사도 :  ',cosine_similarity(ENFJ_최종, SF_최종))
print('ENFJ와 공포와의 유사도 : ',cosine_similarity(ENFJ_최종, 공포_최종))
print('ENFJ와 드라마와의 유사도 : ',cosine_similarity(ENFJ_최종, 드라마_최종))
print('ENFJ와 로맨스와의 유사도 : ',cosine_similarity(ENFJ_최종, 로맨스_최종))
print('ENFJ와 스릴러와의 유사도 : ',cosine_similarity(ENFJ_최종, 스릴러_최종))
print('ENFJ와 액션와의 유사도 : ',cosine_similarity(ENFJ_최종, 액션_최종))
print('ENFJ와 코메디와의 유사도 : ',cosine_similarity(ENFJ_최종, 코메디_최종))
print('ENFJ와 판타지와의 유사도 : ',cosine_similarity(ENFJ_최종, 판타지_최종))
```
