```python
import pandas as pd
import pickle
```

```python
import os
os.chdir('/content/drive/MyDrive/Colab Notebooks')
```

```python
filename = 'MBTI_svm_v2.sav'
text_clf = pickle.load(open(filename, 'rb'))
```

```python
# 판타지
fantasy = pd.read_csv('./Naver_Movie_Review_Combination/Fantasy_Combination.csv', encoding='utf-8', index_col=0)
fantasy['Word'] = fantasy['Word'].astype(str)
fantasy_test = fantasy['Word']
predictions_fantasy = text_clf.predict(fantasy_test)
predictions_fantasy = predictions_fantasy.tolist()
```

```python
# 로맨스
romance = pd.read_csv('./Naver_Movie_Review_Combination/Romance_Combination.csv', encoding='utf-8', index_col=0)
romance['Word'] = romance['Word'].astype(str)
romance_test = romance['Word']
predictions_romance = text_clf.predict(romance_test)
predictions_romance = predictions_romance.tolist()
```

```python
# 호러
horror = pd.read_csv('./Naver_Movie_Review_Combination/Horror_Combination.csv', encoding='cp949', index_col=0)
horror['Word'] = horror['Word'].astype(str)
horror_test = horror['Word']
predictions_horror = text_clf.predict(horror_test)
predictions_horror = predictions_horror.tolist()
```

```python
# SF
SF = pd.read_csv('./Naver_Movie_Review_Combination/SF_Combination.csv', encoding='utf-8', index_col=0)
SF['Word'] = SF['Word'].astype(str)
SF_test = SF['Word']
predictions_SF = text_clf.predict(SF_test)
predictions_SF = predictions_SF.tolist()
```

```python
# 스릴러
thriller = pd.read_csv('./Naver_Movie_Review_Combination/Thriller_Combination.csv', encoding='utf-8', index_col=0)
thriller['Word'] = thriller['Word'].astype(str)
thriller_test = thriller['Word']
predictions_thriller = text_clf.predict(thriller_test)
predictions_thriller = predictions_thriller.tolist()
```

```python
# 코미디
comedy = pd.read_csv('./Naver_Movie_Review_Combination/Comedy_Combination.csv', encoding='utf-8', index_col=0)
comedy['Word'] = comedy['Word'].astype(str)
comedy_test = comedy['Word']
predictions_comedy = text_clf.predict(comedy_test)
predictions_comedy = predictions_comedy.tolist()
```

```python
# 액션
action = pd.read_csv('./Naver_Movie_Review_Combination/Action_Combination.csv', encoding='utf-8', index_col=0)
action['Word'] = action['Word'].astype(str)
action_test = action['Word']
predictions_action = text_clf.predict(action_test)
predictions_action = predictions_action.tolist()
```

```python
# 드라마
drama = pd.read_csv('./Naver_Movie_Review_Combination/Drama_Combination.csv', encoding='utf-8', index_col=0)
drama['Word'] = drama['Word'].astype(str)
drama_test = drama['Word']
predictions_drama = text_clf.predict(drama_test)
predictions_drama = predictions_drama.tolist()
```

```python
def MBTI_Movie_Preference(x):
  Genre_list = ['fantasy','romance','horror','SF','drama','action','thriller','comedy']
  for genre in Genre_list:
    print(x+'-'+'{}'.format(genre)+' : ' + str(round(eval('predictions_'+'{}'.format(genre)).count(x)/(predictions_fantasy.count(x) + predictions_romance.count(x)\
                                                           + predictions_horror.count(x) + predictions_SF.count(x)\
                                                           + predictions_drama.count(x) + predictions_action.count(x)\
                                                           + predictions_thriller.count(x) + predictions_comedy.count(x))*100,2)
                            ) + "%"
  )
```

```python
MBTI_Movie_Preference('INTP')
print('\n')
MBTI_Movie_Preference('INFP')
print('\n')
MBTI_Movie_Preference('ISFP')
print('\n')
MBTI_Movie_Preference('ISTP')
print('\n')
MBTI_Movie_Preference('ENFP')
print('\n')
MBTI_Movie_Preference('ENTP')
print('\n')
MBTI_Movie_Preference('ESFP')
print('\n')
MBTI_Movie_Preference('ESTP')
print('\n')
MBTI_Movie_Preference('ENFJ')
print('\n')
MBTI_Movie_Preference('ENTJ')
print('\n')
MBTI_Movie_Preference('ESFJ')
print('\n')
MBTI_Movie_Preference('ESTJ')
print('\n')
MBTI_Movie_Preference('INFJ')
print('\n')
MBTI_Movie_Preference('INTJ')
print('\n')
MBTI_Movie_Preference('ISFJ')
print('\n')
MBTI_Movie_Preference('ISTJ')
print('\n')
```
